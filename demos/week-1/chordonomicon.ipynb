# Join!!!!

```python
df.spotify_song_id.count() / len(df)
```

```python
# This makes it so we show more columns in the dataframes in the notebook
pd.set_option('display.max_columns', 50)
```

```python
spotify_songs = pd.read_csv("./week-1/spotify_songs.csv")
print(spotify_songs.shape)

spotify_songs.head()
```

```python
# Inner join: we only keep the songs that have spotify IDs in both the chordonomicon dataset AND the spotify dataset
df_with_spotify = pd.merge(df, spotify_songs, left_on='spotify_song_id', right_on='track_id')

print(len(df_with_spotify))

df_with_spotify.head()
```

```python
# If we wanted everything in chordonomicon REGARDLESS of whether it had a spotify record linked
df_with_spotify_left_join = pd.merge(df, spotify_songs, left_on='spotify_song_id', right_on='track_id', how='left')

len(df_with_spotify_left_join), len(df)
```

```python
spotify_songs.track_id.value_counts()
```

Are the songs from the spotify dataset that show up in chordonomicon MORE POPULAR than the songs that don't show up? Is that why they got tabbed?

```python
df_with_spotify.track_popularity.mean(), df_with_spotify.track_popularity.std()
```

```python
spotify_songs.track_popularity.mean(), spotify_songs.track_popularity.std()
```

```python
# Let's get the spotify songs that aren't in chordonomicon:

spotify_songs_not_in_chordonomicon = spotify_songs[
    ~spotify_songs.track_id.isin(df.spotify_song_id)
]

len(spotify_songs_not_in_chordonomicon)
```

```python
spotify_songs_not_in_chordonomicon.track_popularity.mean()
```

It seems like there is something here! The songs in chordonomicon are more popular than the songs that aren't!
